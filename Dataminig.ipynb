# **Handling Missing Values**
Conert all -200 to NaN

Then replace all NaN values with mean of secific column

```python
air_quality_data=air_quality_data.replace(to_replace=-200,value=np.NaN)
```

```python
#checking missing values
air_quality_data.isnull().sum()
```

```python
# Fill missing values with the mean of each column for columns 3 to 15
air_quality_data.iloc[:, 2:] = air_quality_data.iloc[:, 2:].fillna(air_quality_data.iloc[:, 2:].mean())
```

The overall forecast accuracy is 94.57%

```python
#coneverting the date DD/MM/YYYY to YYYY-MM-DD

data_info = pd.to_datetime(air_quality_data['Date'], format='%d/%m/%Y')
print(data_info)

```

```python
time_info=air_quality_data['Time']
```

```python
time_info = time_info.apply(lambda x: x.replace('.', ':'))
print(time_info)
```

```python
print(type(data_info))
print(type(time_info))
```

```python
date_time = pd.concat([data_info,time_info],axis=1)
```

```python
date_time.head()
```

```python
# formatting
date_time['ds']=date_time['Date'].astype(str)+' '+date_time['Time'].astype(str)
date_time.head()
```

```python
data=pd.DataFrame()
```

```python
data['ds']=pd.to_datetime(date_time['ds'])
```

```python
data.head()
```

```python
data['y']=air_quality_data['RH']
```

```python
data.head()
```

```python
pip install prophet
```

```python
from prophet import Prophet

```

```python
m = Prophet()
m.fit(data)
```

```python
future = m.make_future_dataframe(periods=365,freq='H')
future.tail()
```

```python
forecast = m.predict(future)
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()

```

```python
fig1 = m.plot(forecast)

```

```python
fig2 = m.plot_components(forecast)
```

```python
from prophet.plot import plot_plotly, plot_components_plotly

plot_plotly(m, forecast)

```
